## Question 1 
**Following your graduation, you are hired by a polling organisation as a data analyst. As social media has exploded and transformed the way people interact with each other, it would be a great idea to use messages collected from social media to predict how the user can be converted to change his/her support. List three challenges to solving this problem. With reference to existing approaches, describe the design of your system. Discuss the ethical and social consequences of this study.**



From this question, the problem is conceptually a prediction problem, The behavior, social context, interests and education of an individual are factors that influence their political opinion. Therefore, to change an individual's political stance, the first goal is to determine their current political opinion, the largest factors that influence them to adopt their current view and then creating a solution to change their views. 

The first challenge is determining what the users current political view is. On social media platforms like Facebook and Twitter, users are able to make posts or tweets for their opinions. There are multiple approaches to this problem, a possible approach is to analyse a users posts and analyses the use of language regarding a particular policy. In the example of US politics, a Liberal supporter called “estate tax” as opposed to a conservative “death tax” (Pennacchiotti and Ana-Maria, 2011). These choice of words can help determine the user's political disposition. Another possible approach is the use of image recognition to predict their political  stance via their profile image. One of the more controversial political topics is on same sex marriage, where supporters whom affiliate themselves with the left, change their profile picture to incoporate a rainbow filter. Which can be a strong factor in determining a users political position.

The second challenge is determining what factors affect or cause the user to adopt a particular political position. A political stance is generated by what a user personal values align with the closest which can host many factors such as social, educational and behavioral concepts. While a user may not necessarily convey all their details onto a social platform, some data can be extrapolated for use in prediction. In particular, on Twitter, a user can follow another user or append hashtags to their tweets. By examining the hashtags, or their followings, their interest can be found and analysed to determine what topics are commonly tagged for users of the same political position. An example, in a research article by (Mancosu and Bobba, 2019), A Convoluted Neural Network algorithm was used in image recognition to predict their age, mood and gender to determine whether they supported or opposed Brexit.

The last and probably most difficult challenge is the actual persuasion of a users political view. From the previous problem, you could possibly make targeted advertisements for the opposing political group by showing the advantages of the other political stance that aligns with the users most valued values.. But this is entirely dependent on the users personal behaviour. In previous machine learning studies in politics (Nickerson and Rogers, 2013), politicians elected to only use persuasion strategies only on voters who were politically neutral as to gain more votes, as there was no benefit in persuading supporters and the difficulty in persuading opposers was usually too strongly resisted to have an effect. 

To redefine the problem, the goal of this system is to predict which political side they are on, depending on their stance on each issue. Eg. If conservatives adopted gay marriage, which political aligns mostly with these views. Before delving into the design of the system, an appropriate model needs to be chosen by considering the context of the problem where:

*     Binary prediction problem between two political parties. (Keeping it to two for simplicities sake)
*    Data is categorically binary. (Support, Not Support )
*    Useful to know what variables affect the most



Therefore, the best model for this is a decision tree, as we want to know which political standing a user has according to their stances on particular issues. Additionally, a tree is much simpler and allows the polling organization to know which topics have the most effect on influencing a users political disposition. The main problem however is acquiring and preprocessing the data so that it can be used for this prediction. Users first need to examined as to whether they support or not support every political issue and transform that data from a tweet/post into a acceptable data format. However, in reality, it is not likely that a user would post their stance on every political issue which requires a method for handling missing values.

For this machine learning problem, it can be ethically ambiguous as to whether it infringes on the privacy of users or not. In comparison to the Cambridge Analytica Scandal, where users personal details were obtained without their explicit permission. This solution merely looks upon the history of tweets and posts by a user. Although it can also be argued that the use of analytics of their posts without their explicit permission is ethically questionable. The main difference that distinguishes the lack of ethics in machine learning is whether the users are aware that their personal details are being used. As an example, in Obama’s 2012 campaign, where an app was used to collect data from the user, it also had access to the app user’s friend details. Which was ignored and removed in contrast to Cambridge Analytica.
